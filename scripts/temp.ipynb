```python
# WORLD = np.zeros((13,13), dtype=np.int8)
WORLD = generateWarehouse(length=LENGTH)

print(sum(row.count(0) for row in WORLD.tolist()))
STARTS = list()

while(len(STARTS)!=NUM_AGENTS):
    STARTS.append(getFreeCell(STARTS, WORLD))
    
GOALS = list()
while(len(GOALS)!=NUM_AGENTS):
    GOALS.append(getFreeCell(GOALS, WORLD))
WORLD = WORLD.tolist()
paths = lacam.solve(WORLD, STARTS, GOALS, 3)

ACTIONS = []

for i in range(NUM_AGENTS):
    action = []
    for j in range(len(paths[i])-1):
        action.append(getActionFromPos(paths[i][j], paths[i][j+1]))
    ACTIONS.append(action)

idx = max(len(i) for i in ACTIONS)

temp = np.zeros((NUM_AGENTS, idx))
for i in range(len(ACTIONS)):
    for j in range(len(ACTIONS[i])):
        temp[i,j] = ACTIONS[i][j]

ACTIONS = temp
```
